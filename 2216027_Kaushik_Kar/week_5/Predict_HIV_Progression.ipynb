Convert Sequence to Vector encodings

```python
def convert_to_matrixPR(df:pd.DataFrame,max_len:int) -> pd.DataFrame:
    """
    Convert the sequence into set of columns which encode the PR sequence into 1s and 0s based on letter used
    
    """
    df['PR Seq'] = df['PR Seq'].apply(lambda x: list(x))
    df['PR Seq'] = df['PR Seq'].apply(lambda x: np.array([1 if i == 'A' else 0 for i in x]))
    df['PR Seq'] = df['PR Seq'].apply(lambda x: np.pad(x, (0,max_len-len(x)), 'constant', constant_values=(0, 0)))
    
    df = df.join(pd.DataFrame(df['PR Seq'].values.tolist(), index= df.index)) 
    return df

def convert_to_matrixRT(df:pd.DataFrame,max_len:int) -> pd.DataFrame:
    """
    Convert the sequence into set of columns which encode the PR sequence into 1s and 0s based on letter used
    
    """
    df['RT Seq'] = df['RT Seq'].apply(lambda x: list(x))
    df['RT Seq'] = df['RT Seq'].apply(lambda x: np.array([1 if i == 'A' else 0 for i in x]))
    df['RT Seq'] = df['RT Seq'].apply(lambda x: np.pad(x, (0,max_len-len(x)), 'constant', constant_values=(0, 0)))
    df = df.join(pd.DataFrame(df['RT Seq'].values.tolist(), index= df.index))
    return df

df_temp1 = df_raw.copy()
df_temp2 = df_raw.copy()
df_temp1 = convert_to_matrixRT(df_temp1,max_length_RT)
df_temp1.drop(["ResponderStatus","Resp","PR Seq","RT Seq","VL-t0","CD4-t0"],axis=1,inplace=True)

df_temp2 = convert_to_matrixPR(df_temp2,max_length_PR)

df_temp2.drop(["PR Seq","RT Seq"],axis=1,inplace=True)
df_preprocessed = pd.merge(df_temp1,df_temp2,on='PatientID')
print(df_preprocessed.head())
print(df_preprocessed.columns)
columns_req = ["PatientID","Resp","VL-t0","CD4-t0","ResponderStatus"]
for i in columns_req:
    if i not in df_preprocessed.columns:
    
        print(i,"is not in the dataframe")
    else:
        print(i,"is in the dataframe")
```

```python
import tensorflow as tf 
y = df_preprocessed.pop("Resp")
```

```python
#split the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_preprocessed, y, test_size=0.2, random_state=42)

#convert to tensorflow dataset 

batch_size = 32
shuffle_buffer_size = 100
train_dataset = tf.data.Dataset.from_tensor_slices((X_train.values, y_train.values)).shuffle(shuffle_buffer_size).batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test.values, y_test.values)).batch(batch_size)
```

Create tensorflow models

```python
n,c = X_train.shape

normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(np.array(X_train))

inputs = tf.keras.Input(shape=(c,))
x = normalizer(inputs)
x = tf.keras.layers.Dense(128, activation='relu')(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)
model = tf.keras.Model(inputs, outputs)

#binary classification so use binary cross entropy
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(train_dataset, epochs=10,validation_data=test_dataset)
model.save("model.h5")
#predict on the test set
y_pred = model.predict(test_dataset)

num_of_errors = 0
for i in range(len(y_pred)):
    if y_pred[i] > 0.5:
        y_pred[i] = 1
    else:
        y_pred[i] = 0
    if y_pred[i] != y_test.values[i]:
        num_of_errors += 1
print(f"score is {1-(num_of_errors/len(y_pred))}")
```
